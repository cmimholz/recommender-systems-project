```python
# Define the sample size (e.g., 19% of the dataset)
sample_size = 0.15  # This means 19% of the data

# Generate a random sample of the dataset
random_seed = 42
sampled_data = data.sample(frac=sample_size, random_state=random_seed)

# Save the sample to a new CSV file
sampled_file_path = '../data/sampled_data.csv'
sampled_data.to_csv(sampled_file_path, index=False)

print(f'Sampled data saved to {sampled_file_path}')
```

```python
import random
#To create a smaller dataset (15% of the original) while ensuring that if a user is included, all songs listened to by that user are included

# Assuming you already have your dataframe loaded as `data`
# Let's first make sure that we have the correct size of the sample (15%)
new_var = data
total_rows = len(new_var)
sample_size = int(total_rows * 0.15)

# Shuffle the dataframe
data = data.sample(frac=1, random_state=42).reset_index(drop=True)

# Get the unique users to sample from
unique_users = data['user_id'].unique()

# Initialize the sampled dataframe and a set to track added users
sampled_data_complete_user_songs = pd.DataFrame()
added_users = set()

# Continue sampling until the size of the sampled dataframe is 15% of the original
while len(sampled_data_complete_user_songs) < sample_size:
    # Randomly select a user that hasn't been added yet
    user_id = random.choice(unique_users)
    
    if user_id not in added_users:
        # Get all the records for the selected user where the song was listened to (is_listened == 1)
        user_songs = data[(data['user_id'] == user_id) & (data['is_listened'] == 1)]
        
        # Only add the user's songs if there are any that match the condition
        if not user_songs.empty:
            # Append the user's data to the sampled dataframe
            sampled_data_complete_user_songs = pd.concat([sampled_data_complete_user_songs, user_songs])
        
        # Mark this user as added
        added_users.add(user_id)

# If the sampled dataframe exceeds the desired size, trim it down
sampled_data_complete_user_songs = sampled_data_complete_user_songs.head(sample_size)

# Reset the index of the sampled dataframe
sampled_data_complete_user_songs = sampled_data_complete_user_songs.reset_index(drop=True)
```

```python
from sklearn.model_selection import train_test_split

# Assuming `sampled_data_complete_user_songs` is your dataframe
df = sampled_data_complete_user_songs

# First, split into train+validation and test (let's assume 70% for train+validation and 30% for test)
train_val_df, test_df = train_test_split(df, test_size=0.30, random_state=42, stratify=df['is_listened'])

# Now split the train+validation set further into train and validation (let's assume 80% for train and 20% for validation)
train_df, val_df = train_test_split(train_val_df, test_size=0.20, random_state=42, stratify=train_val_df['is_listened'])
```

```python
# Saving each dataframe to a separate CSV file
train_df.to_csv('../data/train_dataset.csv', index=False)
val_df.to_csv('../data/validation_dataset.csv', index=False)
test_df.to_csv('../data/test_dataset.csv', index=False)
```
